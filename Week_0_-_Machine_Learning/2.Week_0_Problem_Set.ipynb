__4. Run the code below. Then, explain why the ML model doesn't give exactly 6 as the answer, even though 4 + 2 = 6.__

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Training data
X = np.array([[1, 1], [2, 2], [3, 1], [5, 3]])
y = np.array([2, 4, 4, 8])

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[4, 2]])[0]
print(f"ML Model says 4 + 2 = {prediction:.4f}")
print(f"Actual answer: 4 + 2 = 6")
```

**Answer 4:**

> Machine learning models use mathematical optimization to find a line of best fit (statistical approximation) based on the training data, rather than learning strict symbolic logic or arithmetic. Because of this approximation and floating-point math, the prediction is a statistical estimate (like 5.9999...) rather than an exact integer 6.

## Part 2: The "Black Box" and Data
Because machine learning models learn patterns rather than following strict human-written rules, they can sometimes be difficult to understand.

__5. Explain in your own words what a "black box" means in the context of machine learning. Give one specific reason why this lack of transparency might be a problem in the real world.__

**Answer 5:**

> A "black box" means the internal decision-making process of the model is hidden or too complex for humans to understand how it arrived at a specific conclusion. This is a problem in the real world because if an AI makes a critical mistake (e.g., denying a loan or misdiagnosing a patient), we cannot easily explain *why* it made that choice, making it difficult to trust, audit, or fix the system.

## Part 3: Understanding Bias
If a model is a "black box," we have to be very careful about the data we put into it. "Garbage in, garbage out" is a common phrase in data science. When the input data is flawed, the model's predictions will be biased.

__6. Historical Bias:__ A company wants to build a machine learning model to predict which job applicants to interview. They plan to train it on 5 years of historical hiring data. Identify TWO potential sources of bias in this approach.

**Answer 6:**

1. **Human bias in past decisions:** If the human recruiters over the last 5 years were biased against certain demographics, the model will learn to replicate those same biased hiring decisions.
2. **Imbalanced training data:** If the historical data contains mostly applicants from a specific demographic (e.g., predominantly men), the model might incorrectly correlate being part of that majority group with being a successful candidate.

__7. Representation Bias:__ A facial recognition system is trained primarily on images of people with light skin. What type of bias is this, and what real-world problems might it cause?

**Answer 7:**

> This is **Representation Bias** (or Sample Bias). Because the training data does not accurately represent the diverse population, the system will perform poorly on people with darker skin tones. In the real world, this can cause issues like false arrests if used by law enforcement, or locking people out of their own devices.

__8. Proxy Bias:__ Consider the following scenario:

A hospital uses a machine learning model to predict which patients need urgent care. The model was trained on historical data where patients from wealthier areas received more tests and therefore had more documented health issues.

a) What type of bias might this model have?
b) How might this affect patients from less wealthy areas?
c) What ethical principle is being violated?

**Answer 8:**

**a)** This is an example of **Proxy Bias** (specifically, measurement bias). The model is using the number of tests/documented health issues as a proxy for actual health risk, which is heavily influenced by wealth and healthcare access.

**b)** Patients from less wealthy areas might be wrongly classified as "lower risk" simply because they couldn't afford or didn't receive as many tests in the past, causing them to be denied urgent care when they actually need it.

**c)** The ethical principle of **Fairness** (and equity) is being violated.

__10. Case Study: Predictive Policing__

A city implements a machine learning system to predict where crimes are likely to occur. The system is trained on historical arrest data from the past 10 years. Police are then sent to patrol areas the model predicts will have high crime.

a) If certain neighborhoods have historically been over-policed (more police presence leading to more arrests), how might this affect the model's predictions?
b) Could this create a feedback loop? Explain how.
c) What questions should be asked before deploying such a system?
d) Who should be accountable if the system leads to unfair outcomes?

**Answer 10:**

**a)** The model will falsely predict that these neighborhoods inherently have higher crime rates, because it is confusing "more arrests due to higher police presence" with "more actual crime."

**b)** Yes, it creates a feedback loop. The model predicts high crime in a neighborhood -> Police are sent there -> Increased police presence leads to more arrests -> These new arrests are fed back into the model -> The model predicts even higher crime there in the future.

**c)** Questions should include: Does the historical data reflect actual crime rates or just policing habits? Are we disproportionately targeting specific minority communities? How will we measure false positives?

**d)** The city officials, the police department deploying the system, and the developers who designed it should all share accountability for auditing the system and mitigating unfair outcomes.
